```python
# Plot functions
def plot_points(x, y, colors = ['blue', 'red'], alpha=1):
    for i in [0,1]:
        plt.scatter(x[y==i,0], x[y==i,1], marker="x", c=colors[i], label=i, alpha=alpha)
    plt.legend(loc="lower right")
    
def plot_contourlines(x, y, model, delta = 50, alpha=0.5):
    x_min = np.min(x,axis=0)                                      
    x_max = np.max(x,axis=0)                                      
    x1_lin = np.linspace(x_min[0], x_max[0], delta)                     
    x2_lin = np.linspace(x_min[1], x_max[1], delta)                     

    MX1, MX2 = np.meshgrid(x1_lin, x2_lin)                              
    x_grid = np.asarray([MX1.flatten(),MX2.flatten()]).T
    
    if hasattr(model, 'predict_proba'):
        s_grid =  model.predict_proba(x_grid)[:,1]
    elif hasattr(model, 'decision_function'):
        s_grid = model.decision_function(x_grid)
    elif hasattr(model, 'predict'):
        s_grid =  model.predict(x_grid)  
        
    CS = plt.contour(x_grid[:,0].reshape(delta,delta),
                     x_grid[:,1].reshape(delta,delta),
                     s_grid[:].reshape(delta,-1), linewidths=3,
                     alpha=alpha)

    plt.clabel(CS)
    plt.grid(True) 
            
def plot_densities(y_pred, y, bandwidth=None, proportion=1.0):
    delta = 100
    s_min = min(y_pred)
    s_max = max(y_pred)
        
    for i in [0, 1]:
        if bandwidth is None:
            bandwidth = (np.abs(s_max-s_min)+1)/20
        #d_estimator = PGMM()
        #d_estimator = VBGMM(n_components=20, alpha=10000)
        #d_estimator = DPGMM(n_components=20, alpha=100, n_iter=100)
        #d_estimator = KernelDensity(kernel='gaussian', bandwidth=bandwidth)
        #d_estimator = GaussianMixtureModel(n_components=10)
        d_estimator = GMM(n_components=10)

        s = y_pred[y==i].reshape(-1,1)
        s_grid = np.linspace(s_min, s_max, delta).reshape(-1,1)

        d_estimator.fit(s)
         
        if type(d_estimator) in [VBGMM]:
            parameters = zip(d_estimator.means_.flatten(), 
                         np.concatenate(d_estimator._get_covars()).flatten(), 
                         d_estimator.weights_)
            de_copy = GaussianMixtureModel(params=parameters)
            s_pred = de_copy.pdf(s_grid)
        elif type(d_estimator) in [GMM]:
            s_pred = np.exp(d_estimator.score_samples(s_grid))
        else:
            s_pred = d_estimator.score_samples(s_grid)
            s_pred = np.exp(s_pred)
        plt.plot(s_grid, s_pred, c=colors[i])
    max_index = int(proportion*len(y_pred))
    plt.scatter(y_pred[:max_index], np.zeros((max_index, 1)), marker='^',
                s=200, c=[colors[int(i)] for i in y[:max_index]], alpha=0.1)
    plt.ylim(0)
    plt.xlabel('Scores')
    plt.ylabel('Estimated density')
    plt.grid(True) 
    #plt.hist([y_pred[y==0], y_pred[y==1]], normed=1, color=colors)

def plot_reliability_diagram(scores_set, labels, legend_set,                    
                             original_first=False, alpha=1, show_legend=True,
                             proportion=1.0,
                             **kwargs):                                                                      
    #ax.set_ylim([0,1])                                                          
    #ax.set_xlim([0,1])                                                          
    n_lines = len(legend_set)                                                   
    if original_first:
        s_min = min(scores_set[0])
        s_max = max(scores_set[0])
        bins = np.linspace(s_min,s_max,11)                                              
        hist_tot = np.histogram(scores_set[0], bins=bins)                       
        hist_pos = np.histogram(scores_set[0][labels == 1], bins=bins)          
        edges = np.insert(bins, np.arange(len(bins)), bins)                     
        empirical_p = np.true_divide(hist_pos[0]+alpha, hist_tot[0]+2*alpha)    
        empirical_p = np.insert(empirical_p, np.arange(len(empirical_p)),       
                empirical_p)                                                    
        plt.plot(edges[1:-1], empirical_p, label='empirical')                    
                                                                                
    skip = original_first                                                       
    for (scores, legend) in zip(scores_set, legend_set):                        
        if skip and original_first:                                             
            skip = False                                                        
        else:                                                                   
            reliability_diagram(scores, labels, marker='x-',                    
                    label=legend, linewidth=n_lines, alpha=alpha, **kwargs)     
            n_lines -= 1                                                        
    if original_first:
        last_index = int(proportion*len(scores_set[0]))
        plt.scatter(scores_set[0][:last_index], labels[:last_index],
                    marker='x',
                    c=[colors[int(i)] for i in labels][:last_index],
                    label=legend_set[0],               
                    s=30, alpha=0.5)                                
    #ax.plot([0,1],[0,1], 'r--')
    if show_legend:
        plt.legend(loc='lower right') 
    plt.xlabel('Scores')
    plt.ylabel('Empirical probability')
    plt.grid(True)      
                                               
                                                                                
class MyDecisionTreeClassifier(DecisionTreeClassifier):                         
    def predict_proba(self,X):                                                  
        alpha = 1                                                               
        leaf_indices = self.apply(X)                                            
        leaf_counts = self.tree_.value[leaf_indices]                            
        leaf_counts = leaf_counts.squeeze()                                     
        leaf_counts += alpha                                                    
        return leaf_counts/leaf_counts.sum(axis=1).reshape(-1,1)
```

```python
np.random.seed(2)
N = 1000

n_subclasses = 10
sparseness = 90
mean = np.array([-1, 1])
cov = np.matrix([[1,0],[0,1]])

def cov_variability(variability=20):
    return np.random.rand()*variability

for i in range(n_subclasses):
    mean_aux = mean + (np.random.rand(2)-0.5)*sparseness
    if i == 0:
        x = np.random.multivariate_normal(mean_aux, cov*cov_variability(), N)
    else:
        x = np.vstack((x, np.random.multivariate_normal(mean_aux, cov*cov_variability(), N)))

for i in range(n_subclasses):
    mean_aux = -mean - (np.random.rand(2)-0.5)*sparseness
    x = np.vstack((x, np.random.multivariate_normal(mean_aux, cov*cov_variability(), N)))

y = np.hstack((np.zeros(N*n_subclasses), np.ones(N*n_subclasses)))
x, y = shuffle(x, y)

plt.figure()
plot_points(x,y, alpha=0.5)
```

```python
from sklearn.calibration import CalibratedClassifierCV
from sklearn.model_selection import StratifiedKFold
#from sklearn.mixture import PGMM
from sklearn.metrics import log_loss

models = [#(LinearRegression, {},"Linear Regression"),
          (LogisticRegression, {'C': 0.0001},"Logistic Regression"),
          (KNeighborsClassifier, {},"K-Nearest Neighbors"),
          (MyDecisionTreeClassifier, {},"Decision Tree"),
          (GaussianNB, {},"Gaussian Naive Bayes"),
          #(SVC, {'kernel':'linear'},"Linear SVM"),
          #(SVC, {'kernel':'poly', 'degree':2},"Polynomial 2 SVM"),
          #(SVC, {'kernel':'poly', 'degree':3},"Polynomial 3 SVM"),
          (SVC, {'kernel':'rbf'},"RBF SVM"),
          (MLPClassifier, {}, "MLP")]
n_models = len(models)

titles = ["Contour lines of Scores", 
          "Density of Scores",
          "Class proportions"]

fig = plt.figure(figsize=(4.5*3, 2*len(models)))
for i, title in enumerate(titles):
    plt.subplot(n_models, 3, i+1)
    plt.title(title)

def get_sets(x, y, test_fold_id, test_folds):                                   
    x_test = x[test_folds == test_fold_id, :]                                   
    y_test = y[test_folds == test_fold_id]                                      
                                                                                
    train_indices = test_folds != test_fold_id                                  
    x_train = x[train_indices, :]                                               
    y_train = y[train_indices]                                                  
                                                                                
    return [x_train, y_train, x_test, y_test]

skf = StratifiedKFold(n_splits=2)  

train_idx, val_idx = next(skf.split(x, y))
x_train = x[train_idx]
y_train = y[train_idx]
x_val = x[val_idx]
y_val = y[val_idx]

for i, (model, args, name) in enumerate(models):
    model = model(**args)
    print("Fitting model {}".format(name))
    print(model.get_params())
    model.fit(x_train,y_train)
    score = model.score(x_val, y_val)
    print("Accuracy = {}".format(score))
      
    if hasattr(model, 'predict_proba'):
        y_val_pred = model.predict_proba(x_val)[:,1]
    elif hasattr(model, 'decision_function'):
        y_val_pred = model.decision_function(x_val)
    else:
        y_val_pred = model.predict(x_val)
        
    logloss = log_loss(y_val, y_val_pred)
    
    # CONTOUR LINES
    plt.subplot(n_models,3,i*3+1)
    plt.ylabel(name + '\n$X_1$')
    plt.xlabel('$X_0$')
    plot_points(x_val[:50],y_val[:50],alpha=1.0)
    plot_contourlines(x,y,model,alpha=1)
    if i < len(models) - 1:
        plt.xlabel('')
    
    # SCORES' DENSITY
    plt.subplot(n_models,3,i*3+2)
    s_min = min(y_val_pred)
    s_max = max(y_val_pred)
    bandwidth = (np.abs(s_max-s_min)+1)/20
    plot_densities(y_val_pred, y_val, bandwidth, proportion=0.2)
    
    if i < len(models) - 1:
        plt.xlabel('')
    
    plt.subplot(n_models,3,i*3+3)
    plot_reliability_diagram((y_val_pred,), 
                     y_val,
                     legend_set=('original',),
                     original_first=True,
                     show_legend=False,
                     proportion=0.2)
    if i < len(models) - 1:
        plt.xlabel('')
        
    
    
plt.tight_layout()
plt.savefig('example_classifier_scores.pdf')
```
